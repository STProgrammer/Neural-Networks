```python
valid_split = 0.2

transform = transforms.Compose([transforms.ToTensor(), transforms.Normalize((0.5, ), (0.5, ))])
train_dataset = datasets.MNIST('Documents\\Programs\\Python\\PythonExercises\\Neural networks\\MNIST_dataset.csv', download=True, train=True, transform=transform)
train_dataset, valid_dataset = torch.utils.data.random_split(train_dataset, [int(round((1-valid_split)*len(train_dataset))), int(round(valid_split*len(train_dataset)))])
test_dataset = datasets.MNIST('Documents\\Programs\\Python\\PythonExercises\\Neural networks\\MNIST_dataset.csv', download=True, train=False, transform=transform)

train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=32, shuffle=True)
valid_loader = torch.utils.data.DataLoader(valid_dataset, batch_size=32, shuffle=True)
test_loader = torch.utils.data.DataLoader(test_dataset, batch_size=32, shuffle=True)
```

```python


images, labels = next(iter(train_loader))
print("First batch images", images.shape)
print("First batch labels", labels.shape)
image = images[0]
label = labels[0]
print("One image shape", image.shape)
print(label)
plt.imshow(image.squeeze().numpy(), cmap='gray')
```

```python
# network
model = nn.Sequential(nn.Linear(784, 256),
                     nn.ReLU(),
                     nn.Linear(256, 64),
                     nn.ReLU(),
                     nn.Linear(64, 10))
```

```python
criterion = nn.CrossEntropyLoss()

optimizer = optim.SGD(model.parameters(), lr = 0.001)
```

```python
num_epochs = 15

x_axis = np.linspace(0, num_epochs, num_epochs)

sgd_results = list()
model.train()
for ep in range(num_epochs):
  running_loss = 0
  for images, labels in train_loader:
    images = images.view(images.shape[0], -1)
    # Transfer Data to GPU if available
    # if torch.cuda.is_available():
    #   images, labels = images.cuda(), labels.cuda()
    optimizer.zero_grad()
    output = model(images)

    loss = criterion(output, labels)
    loss.backward()
    optimizer.step()

    running_loss += loss.item()
  else:
    sgd_results.append(running_loss / len(train_loader))
print(sgd_results)

```

```python
model.eval()
valid_loss = 0
with torch.no_grad():
  for images, labels in valid_loader:
    images = images.view(images.shape[0], -1)
    valid_output = model(images)
    loss = criterion(valid_output, labels)
    valid_loss += loss.item()
  else:
    mean_valid_loss = valid_loss / len(valid_loader)
print(mean_valid_loss)
```

```python

model.eval()
valid_loss = 0
with torch.no_grad():
  for images, labels in test_loader:
    images = images.view(images.shape[0], -1)
    test_output = model(images)
    loss = criterion(valid_output, labels)
    test_loss += loss.item()
  else:
    mean_test_loss = test_loss / len(test_loader)
print(mean_test_loss)
```
