```python
import pandas as pd

dataset = pd.read_csv('iris.csv', header=None, names=['sepal length', 'sepal width', 'petal length', 'petal width', 'class'])

dataset.drop(index=dataset.index[dataset['class'] == 'Iris-setosa'], inplace=True)

dataset.drop('petal length', axis='columns', inplace=True)


binary_dataset = dataset.copy()

binary_dataset.loc[dataset['class'] == 'Iris-versicolor', dataset.columns == 'class'] = 0
binary_dataset.loc[dataset['class'] == 'Iris-virginica', dataset.columns == 'class'] = 1


```
